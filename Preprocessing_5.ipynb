```python
df = df1.append(df2.append(df3.append(df4,ignore_index=True),ignore_index=True),ignore_index=True)
```

```python
df_2021 = df[['Title','Cast and crew']]
```

```python
from tmdbv3api import TMDb
import json
import requests
tmdb = TMDb()
tmdb.api_key = 'cd109cb98c90d3020a252e6f524b8d63'
```

```python
from tmdbv3api import Movie
tmdb_movie = Movie()
def get_movie_id(x):
    try:
        result = tmdb_movie.search(x)
        return result[0].id
    except:
        return np.NaN
```

```python
# based on the movie title , storing the geners for movie
df_2021['id'] = df_2021['Title'].map(lambda x: get_movie_id(str(x)))
```

```python
from tmdbv3api import Movie
tmdb_movie = Movie()
def get_movie_avg_vt(x):
    try:
        result = tmdb_movie.search(x)
        return result[0].vote_average
    except:
        return np.NaN
```

```python
df_2021['vote_average'] = df_2021['Title'].map(lambda x: get_movie_avg_vt(str(x)))
```

```python
from tmdbv3api import Movie
tmdb_movie = Movie()
def get_movie_vt_cnt(x):
    try:
        result = tmdb_movie.search(x)
        return result[0].vote_count
    except:
        return np.NaN
```

```python
df_2021['vote_count'] = df_2021['Title'].map(lambda x: get_movie_vt_cnt(str(x)))
```

```python
from tmdbv3api import Movie
tmdb_movie = Movie() 
def get_genre(x):
    genres = []
    result = tmdb_movie.search(x)
    if not result:
      return np.NaN
    else:
      movie_id = result[0].id
      response = requests.get('https://api.themoviedb.org/3/movie/{}?api_key={}'.format(movie_id,tmdb.api_key))
      data_json = response.json()
      if data_json['genres']:
          genre_str = " " 
          for i in range(0,len(data_json['genres'])):
              genres.append(data_json['genres'][i]['name'])
          return genre_str.join(genres)
      else:
          return np.NaN
```

```python
df_2021['genres'] = df_2021['Title'].map(lambda x: get_genre(str(x)))
```

```python
df_2021
```

```python
def get_director(x):
    if " (director)" in x:
        return x.split(" (director)")[0]
    elif " (directors)" in x:
        return x.split(" (directors)")[0]
    else:
        return x.split(" (director/screenplay)")[0]
```

```python
df_2021['director_name'] = df_2021['Cast and crew'].map(lambda x: get_director(str(x)))
```

```python
def get_actor1(x):
    return ((x.split("screenplay); ")[-1]).split(", ")[0])
```

```python
df_2021['actor_1_name'] = df_2021['Cast and crew'].map(lambda x: get_actor1(str(x)))
```

```python
def get_actor2(x):
    if len((x.split("screenplay); ")[-1]).split(", ")) < 2:
        return np.NaN
    else:
        return ((x.split("screenplay); ")[-1]).split(", ")[1])
```

```python
df_2021['actor_2_name'] = df_2021['Cast and crew'].map(lambda x: get_actor2(str(x)))
```

```python
def get_actor3(x):
    if len((x.split("screenplay); ")[-1]).split(", ")) < 3:
        return np.NaN
    else:
        return ((x.split("screenplay); ")[-1]).split(", ")[2])
```

```python
df_2021['actor_3_name'] = df_2021['Cast and crew'].map(lambda x: get_actor3(str(x)))
```

```python
df_2021.head()
```

```python
df_2021 = df_2021.rename(columns={'Title':'movie_title'})
```

```python
new_df21 = df_2021.loc[:,['id','director_name','actor_1_name','actor_2_name','actor_3_name','genres','movie_title','vote_count','vote_average']]
```

```python
new_df21
```

```python
# replacing the null values with the unknown
new_df21['actor_2_name'] = new_df21['actor_2_name'].replace(np.nan, 'unknown')
new_df21['actor_3_name'] = new_df21['actor_3_name'].replace(np.nan, 'unknown')
```

```python
new_df21['movie_title'] = new_df21['movie_title'].str.lower()
```

```python
new_df21['comb'] = new_df21['actor_1_name'] + ' ' + new_df21['actor_2_name'] + ' '+ new_df21['actor_3_name'] + ' '+ new_df21['director_name'] +' ' + new_df21['genres']
```

```python
new_df21.isna().sum()
```

```python
new_df21 = new_df21.dropna(how='any')
```

```python
new_df21.isna().sum()
```

```python
new_df21['movie_title'] = new_df21['movie_title'].str.lower()
```

```python
new_df21
```

```python
old_df = pd.read_csv('data20.csv')
```

```python
till_2021_df = old_df.append(new_df21,ignore_index=True)
```
