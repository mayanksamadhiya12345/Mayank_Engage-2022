```python
from tmdbv3api import Movie
tmdb_movie = Movie()
def get_movie_avg_vt(x):
    try:
        result = tmdb_movie.search(x)
        return result[0].vote_average
    except:
        return np.NaN
```

```python
df_2020['vote_average'] = df_2020['Title'].map(lambda x: get_movie_avg_vt(str(x)))
```

```python
from tmdbv3api import Movie
tmdb_movie = Movie()
def get_movie_vt_cnt(x):
    try:
        result = tmdb_movie.search(x)
        return result[0].vote_count
    except:
        return np.NaN
```

```python
df_2020['vote_count'] = df_2020['Title'].map(lambda x: get_movie_vt_cnt(str(x)))
```

```python
# by the use of this function , we are extracting the movie related information based on their movie ID with the help of API KEY
from tmdbv3api import Movie
tmdb_movie = Movie() 
def get_genre(x):
    genres = []
    result = tmdb_movie.search(x)
    if not result:
      return np.NaN
    else:
      movie_id = result[0].id
      response = requests.get('https://api.themoviedb.org/3/movie/{}?api_key={}'.format(movie_id,tmdb.api_key))
      data_json = response.json()
      if data_json['genres']:
          genre_str = " " 
          for i in range(0,len(data_json['genres'])):
              genres.append(data_json['genres'][i]['name'])
          return genre_str.join(genres)
      else:
          return np.NaN
```

```python
# based on the movie title , storing the geners for movie
df_2020['genres'] = df_2020['Title'].map(lambda x: get_genre(str(x)))
```

```python
# extracting movie director name 
def get_director(x):
    if " (director)" in x:
        return x.split(" (director)")[0]
    elif " (directors)" in x:
        return x.split(" (directors)")[0]
    else:
        return x.split(" (director/screenplay)")[0]
```

```python
df_2020['director_name'] = df_2020['Cast and crew'].map(lambda x: get_director(str(x)))
```

```python
# extracting movie actror1 name 
def get_actor1(x):
    return ((x.split("screenplay); ")[-1]).split(", ")[0])
```

```python
df_2020['actor_1_name'] = df_2020['Cast and crew'].map(lambda x: get_actor1(str(x)))
```

```python
# extracting movie actror2 name 
def get_actor2(x):
    if len((x.split("screenplay); ")[-1]).split(", ")) < 2:
        return np.NaN
    else:
        return ((x.split("screenplay); ")[-1]).split(", ")[1])
```

```python
df_2020['actor_2_name'] = df_2020['Cast and crew'].map(lambda x: get_actor2(str(x)))
```

```python
# extracting movie actror3 name 
def get_actor3(x):
    if len((x.split("screenplay); ")[-1]).split(", ")) < 3:
        return np.NaN
    else:
        return ((x.split("screenplay); ")[-1]).split(", ")[2])
```

```python
df_2020['actor_3_name'] = df_2020['Cast and crew'].map(lambda x: get_actor3(str(x)))
```
